```python
from abc import ABC, abstractmethod
from numba import njit
class electronic_device:
    def __init__(self, name, **kwargs):
        self.name = name
        print(f'Creating electronic device {self.name}')

    @abstractmethod
    def reset(self):
        print(f'Reseting {self.name}')
    __reset = reset

    def update_firmware(self):
        print(f'Updating firmware of {self.name}')
        self.__reset()
        self.reset()

class computer(electronic_device):
    def __init__(self,ram,**kwargs):
        super().__init__(**kwargs)
        self.ram = ram
        print(f'Creating computer with {self.ram} RAM')
    def reset(self):
        print(f'Saving files and reseting computer {self.name} with {self.ram} RAM')
        super().reset()

class camera(electronic_device):
    def __init__(self, width, height, **kwargs):
        super().__init__(**kwargs)
        self.width = width
        self.height = height
        print(f'Creating camera with {self.width}x{self.height} resolution')
    def reset(self):
        print(f'Saving photos and reseting camera {self.name} with {self.width}x{self.height} resolution')
        super().reset()

class smartphone(camera, computer):
    def __init__(self, name, ram, width, height):
        args = {'name': name, 'ram': ram, 'width': width, 'height': height}
        super().__init__(**args)
        #computer.__init__(self, name, ram)
        #camera.__init__(self, name, width, height) tak nie
        print(f'Creating smartphone with {self.ram} RAM and {self.width}x{self.height} resolution')
    def reset(self):
        print(f'Saving files and photos and reseting smartphone {self.name} with {self.ram} RAM and {self.width}x{self.height} resolution')
        super().reset()
```
