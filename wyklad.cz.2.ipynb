```python
from typing import NamedTuple


class Student(NamedTuple):
    name: str
    age: int
    grade: int
    email: str
    classes: list[str]

bob = Student(name='Bob',age=21,grade = 5, email='123@gmail.com',classes = ['polski','angielski','matematyka'])

print(bob)

print(bob[0])
print(bob[1])
print(bob[3])

print(bob.name)
print(bob.classes)

#bob.name = 'Slawek'   <----NIE DZIALA

```

```python
class Student(NamedTuple):
    name: str = 'Michal'
    age: int = 24
    grade: int = 20
    email: str = 'twzrst123@gmail.com'
    classes: list[str] = []
bob = Student()
alicja = Student(name = 'Alicja', classes = ['matma','fizyka','chemia'])
print(bob)
print(alicja)

bob.classes.append('geografia')
print(bob)
print(alicja)
```

```python
from dataclasses import dataclass, field

@dataclass(slots=True)
class Student:
    name: str 
    age: int 
    grade: int
    email: str

twr = Student(name = 'bob', age = 21, grade = 3, email = 'iutfi.onet.pl')
print(twr)


```

```python
@dataclass
class Student:
    name: str 
    age: int 
    grade: int
    email: str
    classes: list[str] = field(default_factory=list)

```
